```julia
function show_image(path)
    img = load(path)
    h, w = size(img)
    # Create a figure with the same size as the image
    f = Figure(size = (w, h))
    ax = Axis(f[1, 1])  # create axis and hide axis
    hidedecorations!(ax)
    hidespines!(ax)
    ax.aspect = DataAspect()  # maintain image aspect ratio

    image!(ax, rotr90(img))

    f
end
```
